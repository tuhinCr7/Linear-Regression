Important Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading the CSV data file 

```python
car_data=pd.read_csv('/Users/tuhinalambijoy/Documents/Data Science Course Material/car price/train-data.csv')
car_data
```

Cheaking information

```python
car_data.info()
```

Droping the colum

```python
car_data=car_data.drop(columns='Unnamed: 0')
```

Cheaking is there any null value or not

```python
car_data.isnull().sum()
```

Deleting the duplicates values

```python
car_data=car_data.drop_duplicates()
```

```python
car_data
```

In mileage colum after the numeric value there are some string like km/kg and kmpl , removing those using right strip

```python
car_data['Mileage']=car_data['Mileage'].str.rstrip('km/kg')
```

```python
car_data['Mileage']=car_data['Mileage'].str.rstrip('kmpl').astype(float)
```

Also for the Engine colum and making them float 

```python
car_data['Engine']=car_data['Engine'].str.rstrip('CC').astype(float)
```

```python
car_data['Power']=car_data['Power'].str.rstrip('bhp')
```

```python
car_data['Power'].value_counts()
```

```python
car_data
```

cheaking again there is null values or not

```python
car_data.isnull().sum()
```

we can see that in new price colum there are more then 5k values are null among 6k ,so I'm dropping this colum

```python
car_data=car_data.drop(columns='New_Price')
```

Now i'm using dropna becouse there are some null values left in others colum as well as but the number is not much 

```python
car_data=car_data.dropna()
car_data
```

after deleting the null values the index is not same so reseting the index

```python
car_data=car_data.reset_index()
```

in power colum there are some 'null 'value so that i decided to first remove the extra space

```python
car_data['Power'] = car_data['Power'].str.strip()
```

Replacing the 'null' using np,nan

```python
car_data['Power'] = car_data['Power'].replace('null', np.nan)
```

so where the values are 'null' it's become null 

```python
car_data.isnull().sum()
```

deleteing those null values

```python
car_data=car_data.dropna()
```

Now changing the power colum to float

```python
car_data['Power']=car_data['Power'].astype(float)
```

```python
car_data
```

there is extra index so removing that

```python
car_data=car_data.drop(columns='index')
```

```python
car_data.info()
```

The name colum contain so many thing's like company name and brande name , so split them first

```python
car_data[['Company','Brand']]=car_data['Name'].str.split(' ',n=1,expand=True)
```

```python
car_data
```

I wan't only company name so removing the Brand Name

```python
car_data=car_data.drop(columns='Name')
```

```python
car_data=car_data.drop(columns='Brand')
```

```python
car_data
```

```python
car_data['Fuel_Type'].value_counts()
```

Doing One Hot Encoding

```python
car_data = pd.get_dummies(car_data, columns=['Fuel_Type'], drop_first=True)
```

```python
car_data = pd.get_dummies(car_data, columns=['Owner_Type'], drop_first=True)
```

```python
car_data = pd.get_dummies(car_data, columns=['Transmission'], drop_first=True)
```

```python
car_data = pd.get_dummies(car_data, columns=['Location'], drop_first=True)
```

```python
car_data
```

```python
car_data.info()
```

company has so many type so decided to delete this colum

```python
car_data=car_data.drop(columns='Company')
```

```python
car_data.info()
```

Separating Target and feature

```python
X=car_data.drop(columns='Price')
y=car_data['Price']
```

```python
X
```

```python
y
```

Spliting train and test by 80% traing and 20% test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train
```

```python
y_train
```

Linear regression

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
print("Accuracy on Traing set: ",linear_reg.score(X_train,y_train))
print("Accuracy on Testing set: ",linear_reg.score(X_test,y_test))
```

Lesso regression

```python
from sklearn import linear_model

lasso_reg = linear_model.Lasso(alpha=40, tol=0.5)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test,y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X_train, y_train)
```

```python
print("Accuracy on Traing set: ",dt_reg.score(X_train,y_train))
print("Accuracy on Testing set: ",dt_reg.score(X_test,y_test))
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
```

```python
print("Accuracy on Traing set: ",rf_reg.score(X_train,y_train))
print("Accuracy on Testing set: ",rf_reg.score(X_test,y_test))
```
